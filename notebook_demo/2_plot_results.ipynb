# Main - Year long sliding window analysis

```python
STDP = np.load('../data/results/main.npy')[8:24,:,:,:] #border exclusion
print(STDP.shape)
```

```python
estimation=STDP.mean(axis=-1).mean(axis=-1)
estimation[estimation!=0]=1

# estimation bool for the sliding window
plt.imshow(estimation)
plt.show()

full_window = np.where(estimation.mean(axis=0)==1)[0]
rmse_stdp=np.zeros(STDP.shape[1])
rmse_oi=np.zeros(STDP.shape[1])
```

```python
for i in full_window:
    rmse_oi[i]=rmse(OI[i,:,:], Truth[i,:,:])
    rmse_stdp[i]=rmse(STDP.mean(axis=0)[i,:,:], Truth[i,:,:])
```

```python
plt.figure(figsize=(8, 4))

plt.plot(rmse_oi[(32-1):-(32-1)],color='red',linewidth=1,label='DUACS OI')
plt.plot(rmse_stdp[(32-1):-(32-1)],color='blue',linewidth=1, label='Deep prior')

plt.grid()
#plt.ylim([0.015, 0.14])
plt.ylabel('RMSE')
plt.xlabel('time (day)')
plt.legend()


plt.savefig('../figures/year_long.pdf',bbox_inches='tight')
plt.show()
```
