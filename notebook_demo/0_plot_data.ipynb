```python
import numpy as np
import matplotlib.pyplot as plt
import os

#if 'Obs' not in os.listdir('../figures'):
#    os.mkdir('../figures/Obs')
```

# Load

```python
Truth = np.load('../data/ground_truth/ground_truth.npy')[:,-1-128:-1,0:128]
Obs = np.load('../data/observations/ssh_tracks.npy')[:,-1-128:-1,0:128]
Mask = (-1*np.load('../data/observations/mask.npy')+1)[:,-1-128:-1,0:128]
Oi = np.load('../data/estimate/OI/oi_1year.npy')[:,-1-128:-1,0:128]
                           

i = 0
T = 32
sub = 4

truth = Truth[i:i+T,:,:]
obs = Obs[i:i+T,:,:]
mask = Mask[i:i+T,:,:]
oi = Oi[i:i+T,:,:]
```

# Ground truth

```python
vmin = truth.min()
vmax = truth.max()

plt.figure(figsize=(100, 100))

for t in range(T):
    
    if t%sub==0:
        
        plt.subplot(1,int(T/sub),min(int(t/sub)+1,int(T/sub)))
        plt.imshow(truth[i+t,:,:],vmin=vmin,vmax=vmax)
        plt.title('day'+str(t), fontsize=75, pad=25)
        #plt.colorbar()
        plt.yticks([])
        plt.xticks([])
    
#plt.savefig('../figures/ground_truth.pdf',bbox_inches='tight')
```
